```python
import csv

# Function to extract words from a CSV file with words in separate columns
def extract_words_from_csv(input_file):
    word_list = []

    with open(input_file, 'r') as input_csv:
        reader = csv.reader(input_csv)
        for row in reader:
            for cell in row:
                # Append each cell value (word) to the list
                word_list.append(cell.strip())  # Use strip() to remove leading/trailing whitespace

    return word_list

if __name__ == "__main__":
    input_file = '/content/drive/MyDrive/indian_names.csv'  # Replace with the path to your input file

    words = extract_words_from_csv(input_file)
    print(words)
```
